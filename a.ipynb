```python
import os
import requests
from pathlib import Path

# Configuration
STATIC_DIR = "static"
PLANE_IMG_PATH = Path(STATIC_DIR) / "plane.jpg"
TRAIN_IMG_PATH = Path(STATIC_DIR) / "train.jpg"
HTML_FILE_PATH = "index.html"

# Image URLs (using Unsplash's public domain images)
IMAGE_URLS = {
    "plane": "https://images.unsplash.com/photo-1590497554904-8933c166d488",
    "train": "https://images.unsplash.com/photo-1544415410-b9846b4122d2"
}

def setup_directories():
    """Create necessary directories if they don't exist"""
    try:
        Path(STATIC_DIR).mkdir(exist_ok=True)
        print(f"Directory ready: {STATIC_DIR}")
    except OSError as e:
        print(f"Error creating directory: {e}")
        raise

def download_image(url, save_path):
    """Download an image from URL and save to specified path"""
    try:
        response = requests.get(url, stream=True, timeout=10)
        response.raise_for_status()
        
        with open(save_path, 'wb') as f:
            for chunk in response.iter_content(chunk_size=8192):
                f.write(chunk)
        print(f"Image saved: {save_path}")
        return True
    except requests.exceptions.RequestException as e:
        print(f"Failed to download image from {url}: {e}")
        return False

def generate_html():
    """Generate HTML file with the downloaded images"""
    html_template = f"""
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Transportation Images</title>
    <style>
        body {{
            font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
            line-height: 1.6;
            color: #333;
            max-width: 1200px;
            margin: 0 auto;
            padding: 20px;
            background-color: #f9f9f9;
        }}
        header {{
            text-align: center;
            margin-bottom: 30px;
        }}
        h1 {{
            color: #2c3e50;
        }}
        .gallery {{
            display: grid;
            grid-template-columns: repeat(auto-fit, minmax(300px, 1fr));
            gap: 20px;
        }}
        .card {{
            background: white;
            border-radius: 8px;
            overflow: hidden;
            box-shadow: 0 4px 6px rgba(0,0,0,0.1);
            transition: transform 0.3s ease;
        }}
        .card:hover {{
            transform: translateY(-5px);
        }}
        .card img {{
            width: 100%;
            height: 200px;
            object-fit: cover;
        }}
        .card-content {{
            padding: 15px;
        }}
        footer {{
            text-align: center;
            margin-top: 40px;
            color: #7f8c8d;
            font-size: 0.9em;
        }}
    </style>
</head>
<body>
    <header>
        <h1>Transportation Gallery</h1>
        <p>Sample images of different transportation methods</p>
    </header>

    <div class="gallery">
        <div class="card">
            <img src="{PLANE_IMG_PATH}" alt="Airplane">
            <div class="card-content">
                <h2>Airplane</h2>
                <p>Commercial passenger airplane flying in the sky</p>
            </div>
        </div>

        <div class="card">
            <img src="{TRAIN_IMG_PATH}" alt="Train">
            <div class="card-content">
                <h2>Train</h2>
                <p>High-speed train moving through the countryside</p>
            </div>
        </div>
    </div>

    <footer>
        <p>Images downloaded from Unsplash | Generated by Python</p>
    </footer>
</body>
</html>
    """
    
    try:
        with open(HTML_FILE_PATH, 'w', encoding='utf-8') as f:
            f.write(html_template)
        print(f"HTML file generated: {HTML_FILE_PATH}")
    except IOError as e:
        print(f"Error writing HTML file: {e}")

def main():
    print("Starting image download and HTML generation...")
    
    try:
        # Setup directory structure
        setup_directories()
        
        # Download images
        download_success = all([
            download_image(IMAGE_URLS["plane"], PLANE_IMG_PATH),
            download_image(IMAGE_URLS["train"], TRAIN_IMG_PATH)
        ])
        
        if not download_success:
            print("Warning: Some images failed to download")
        
        # Generate HTML
        generate_html()
        
        print("\nOperation completed successfully!")
        print(f"You can open '{HTML_FILE_PATH}' in your browser")
        
    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    main()
```
